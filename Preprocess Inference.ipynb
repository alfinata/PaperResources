# Library and Variables

```python
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
from numpy import asarray
import numpy as np
from os import path
import os
```

```python
# Conversion Function
def convertImagefromPath(path, palette, outpath):
  print(path)
  blankpaper = np.ndarray((512, 128, 3), dtype=np.uint8)
  target = np.array(Image.open(path))
  for i, v in enumerate(target):
    for j, w in enumerate(v):
      tmp = w
      blankpaper[i][j] = palette[tmp]
  Image.fromarray(blankpaper).save(f'/content/output/{outpath}')
```

```python
# Initialize Template
blankpaper = np.ndarray((512, 128, 3), dtype=np.uint8)
```

```python
# Conversion Palette
palette = np.array([
  [176, 230, 13],
  [0, 151, 219],
  [126, 230, 225],
  [166, 55, 167],
  [230, 157, 180],
  [167, 110, 77],
  [121, 0, 24],
  [56, 65, 184],
  [230, 218, 0],
  [230, 114, 35],
  [13, 187, 62],
  [230, 182, 22],
  [255, 255, 255]
], dtype=np.uint8)
```
